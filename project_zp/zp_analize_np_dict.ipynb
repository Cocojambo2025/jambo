```python
def write_xls(temp_zp, check_item):
    # Создаем объект ExcelWriter
    with pd.ExcelWriter(check_item+'.xlsx', engine='xlsxwriter') as writer:
        temp_zp.columns = ['Артикул','Наименование', 'Используется в моделях техники', 'Цена \n( с НДС )','Цена со скидкой \n( с НДС )','Размер скидки']
        # Сохраняем DataFrame в Excel
        temp_zp.to_excel(writer, sheet_name='Лист1', index=False)
        
        # Получаем объект workbook и worksheet
        workbook = writer.book
        worksheet = writer.sheets['Лист1']
        
        # Создаем формат числа с разделителем тысяч и двумя знаками после запятой
        number_format = workbook.add_format({'num_format': '#,##0.00'})
        procent_format = workbook.add_format({'num_format': '0%'})
        
        # Применяем формат к нужному столбцу
        # Определяем диапазон столбца (например, столбец A)
        first_row = 1  # первая строка с данными
        last_row = len(temp_zp)+1  # последняя строка
        
        
        # Если нужно применить формат только к определенным строкам:
        worksheet.conditional_format(f'D{first_row}:E{last_row}', 
                                    {'type': 'no_blanks',
                                    'format': number_format})
        worksheet.conditional_format(f'F{first_row}:F{last_row}', 
                                    {'type': 'no_blanks',
                                    'format': procent_format})
        worksheet.set_column('A:A', 10)
        worksheet.set_column('B:B', 50)
        worksheet.set_column('C:C', 15)
        worksheet.set_column('D:D', 12)
        worksheet.set_column('E:E', 12)
        worksheet.set_column('F:F', 8)
        
        header_format = {
        'bold': True,  # Жирный шрифт
        'align': 'center',  # Выравнивание по центру
        'valign': 'vcenter',  # Вертикальное выравнивание
        'bg_color': "#AADB1E",  # Нежно-голубой цвет
        'font_color': 'black',  # Цвет текста
        'border': 1,  # Граница
        'text_wrap': True  # Включаем перенос текста
        }
        
        # Создаем формат заголовка
        header_cell_format = workbook.add_format(header_format)
        
        # Применяем формат к заголовкам
        for col_num, value in enumerate(temp_zp.columns.values):
            worksheet.write(0, col_num, value, header_cell_format)
        
        worksheet.freeze_panes(1, 0)
        
        # Получаем количество столбцов
        num_cols = len(temp_zp.columns)
        last_col = chr(64 + num_cols)  # Преобразуем число в букву столбца
        worksheet.autofilter(f'A1:{last_col}1')
        worksheet.set_row(0, 50)  # 30 - высота в пунктах
```

```python
def check_model(value):
    #if str(value).find(check_item)>-1:
    for keyword in check_item:
        if keyword.upper() in str(value).upper(): 
            return True
    return False
    
def prepare_df(zp):
    temp_zp = zp

    temp_zp['status_model']=temp_zp['Model'].apply(check_model)

    a = temp_zp[temp_zp['status_model']==False].index
    temp_zp = temp_zp.drop(a)

    temp_zp = temp_zp.drop(['status_model'], axis=1)
    return temp_zp
```

```python
def check_model_other(value):
    for element in modeles_dict:
        for keyword in modeles_dict[element]:
            if keyword.upper() in str(value).upper(): 
                return False
    return 

def prepare_df_other(zp):
    temp_zp = zp

    temp_zp['status_model']=temp_zp['Model'].apply(check_model_other)

    a = temp_zp[temp_zp['status_model']==False].index
    temp_zp = temp_zp.drop(a)
    temp_zp = temp_zp.drop(['status_model'], axis=1)
    return temp_zp
```

```python
for element in modeles_dict:
    check_item = modeles_dict[element]
    temp_zp = prepare_df(zp)
    temp_zp = temp_zp.sort_values('Item', ascending=True)
    write_xls(temp_zp, element)
```
